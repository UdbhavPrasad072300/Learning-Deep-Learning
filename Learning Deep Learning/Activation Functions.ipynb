# Activation Functions

List of Activation Functions

```python
import torch
a = torch.Tensor(1, 2, 3)
```

## Sigmoid

$f(x) = \frac{1}{1+e^{-x}} $

Converts data to between 0 to 1 (used to interpret Probability of an Event)<br>
It blows up the values to the extremes <br>
<b>Mostly used in Last Layer, as this can cause the vanishing gradient problem/exploding gradient problem in NNs</b>

```python
#help(torch.sigmoid(a))
```

## Tanh

$f(x) = tanh(x) = \frac{e^{x} - e^{-x}}{e^x + e^{-x}}$

## ReLU 

$f(x) = max(0,x)$

Good for clipping negative values to 0 <br>
Probably the most important one <br>
Here is we get the dying ReLU Problem

```python
#help(torch.nn.ReLU(a))
```

## PReLU

$f(x) = max(x,ax)$

Solves dying ReLU Problem

```python
#help(torch.nn.PReLU(a))
```
